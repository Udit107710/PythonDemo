
# [Linear Regression](https://en.wikipedia.org/wiki/Linear_regression)




```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('http://www-bcf.usc.edu/~gareth/ISL/Advertising.csv',index_col = 0)
print(data.sample(10))                                 ## Taking a look at the dataset
data.shape                                             ## Dimensions of the dataset
```

```python
sns.pairplot(data, x_vars = ['TV','radio','newspaper'], y_vars = 'sales',size = 7, aspect =0.7,kind = 'reg')
```

```python
feature_cols = ['TV','radio','newspaper']

X = data[feature_cols]                                 ## Feature vector
y = data['sales']                                      ## Label vector

print("X: ",X.shape)
print("y: ",y.shape)
```

#### Using train_test_split we'll split the dataset into training and testing data

```python
from sklearn.cross_validation import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y,random_state = 1)
```

##### Using Linear Regression 
Now, we'll import and use Linear Regression model from sklearn. 

```python
from sklearn.linear_model import LinearRegression

lineareg = LinearRegression()

lineareg.fit(X_train,y_train)
```

```python
print(lineareg.intercept_)
print(lineareg.coef_)
```

```python
zip(feature_cols, lineareg.coef_)
```

```python
y_predict = lineareg.predict(X_test)
```

```python
from sklearn import metrics
```

```python
import numpy as np
```

```python
print(np.sqrt(metrics.mean_squared_error(y_test,y_predict)))
```

```python
##Testing after removing newspaper from the dataset

feature_cols = ['TV','radio']

X = data[feature_cols]
y = data['sales']

print(X.shape)
print(y.shape)

from sklearn.cross_validation import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y,random_state = 1)

lineareg = LinearRegression()

lineareg.fit(X_train,y_train)

y_predict = lineareg.predict(X_test)
print(np.sqrt(metrics.mean_squared_error(y_test,y_predict)))
```
