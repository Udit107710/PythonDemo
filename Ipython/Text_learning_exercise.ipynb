```python
import pandas as pd
yelp = pd.read_csv('../datasets/yelp.csv')
yelp.head()
```

```python
#feature matrix and response matrix
X = yelp.text
y = yelp.stars
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)

print (X_train.shape)
print (y_train.shape)
print (X_test.shape)
print (y_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer()
vect.fit(X_train)
vect.get_feature_names()
```

```python
X_train_dtm = vect.transform(X_train)
X_test_dtm = vect.transform(X_test)

print (X_train_dtm.shape)
print (X_test_dtm.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(X_train_dtm, y_train)
```

```python
y_pred_class = nb.predict(X_test_dtm)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred_class)
```

```python
metrics.confusion_matrix(y_test, y_pred_class, labels = [1,2,3,4,5])
```

```python
pos = nb.feature_count_
pos
```

```python
tokens = pd.DataFrame({'tokens': vect.get_feature_names(), '1':pos[0,:], '2':pos[1,:], '3':pos[2], '4':pos[3,:], '5':pos[4,:]})
tokens = tokens.set_index('tokens')
tokens
```

```python
pos = nb.class_count_

tokens['1'] += 1
tokens['2'] += 1
tokens['3'] += 1
tokens['4'] += 1
tokens['5'] += 1
```

```python
tokens['1_freq'] = tokens['1'] / pos[0]
tokens['2_freq'] = tokens['2'] / pos[1]
tokens['3_freq'] = tokens['3'] / pos[2]
tokens['4_freq'] = tokens['4'] / pos[3]
tokens['5_freq'] = tokens['5'] / pos[4]
```
