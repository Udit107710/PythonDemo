```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics as sat
import seaborn as sns
%matplotlib inline
```

```python
train = pd.read_csv('../datasets/LoanTrainData.csv')
test_data = pd.read_csv('../datasets/LoanTestData.csv')

train
```

```python
#converting data to numerical form

train.Gender.replace(('Male','Female'), (1,0), inplace= True)
train.Married.replace(('Yes','No'), (1,0), inplace= True)
train.Education.replace(('Graduate','Not Graduate'), (1,0), inplace=True)
train.Dependents.replace(('0','1','2','3+'),(0,1,2,3), inplace= True)
train.Self_Employed.replace(('Yes','No'), (1,0), inplace=True)
train.Property_Area.replace(('Urban','Semiurban','Rural'),(2,1,0), inplace= True)
train.Loan_Status.replace(('Y','N'),(1,0), inplace= True)
```

```python
#Dealing with NaN 

train.LoanAmount = train.LoanAmount.fillna(train.LoanAmount.mean())
train.Dependents = train.Dependents.fillna(train.Dependents.median())
train.Education = train.Education.fillna(train.Education.median())
train.Self_Employed = train.Self_Employed.fillna(train.Self_Employed.median())
train.ApplicantIncome = train.ApplicantIncome.fillna(train.ApplicantIncome.mean())
train.CoapplicantIncome = train.CoapplicantIncome.fillna(train.CoapplicantIncome.mean())
train.Loan_Amount_Term = train.Loan_Amount_Term.fillna(train.Loan_Amount_Term.median())
train.Credit_History = train.Credit_History.fillna(train.Credit_History.median())
train.Gender = train.Gender.fillna(train.Gender.median())
train.Married = train.Married.fillna(train.Married.median())
train.Property_Area = train.Property_Area.fillna(train.Property_Area.median())
```

```python
# Dealing with outliners 

train.ApplicantIncome.drop(train[train.ApplicantIncome > (train.ApplicantIncome.mean() + train.ApplicantIncome.std() * 1.96)].index)
train.ApplicantIncome.drop(train[train.ApplicantIncome < (train.ApplicantIncome.mean() - train.ApplicantIncome.std() * 1.96)].index)

train.CoapplicantIncome.drop(train[train.CoapplicantIncome > (train.CoapplicantIncome.mean() + train.ApplicantIncome.std() * 1.96)].index)
train.CoapplicantIncome.drop(train[train.CoapplicantIncome < (train.CoapplicantIncome.mean() - train.ApplicantIncome.std() * 1.96)].index)

train.LoanAmount.drop(train[train.LoanAmount > (train.LoanAmount.mean() + train.LoanAmount.std() * 1.96)].index)
train.LoanAmount.drop(train[train.LoanAmount < (train.LoanAmount.mean() - train.LoanAmount.std() * 1.96)].index)
```

```python
#Analysis of relation between gender and Loan_Status


def calculator(new_column):
    if len(new_column.unique()) == 2:
        var_1 = 0;

        for x in new_column:
            if x == 1:
                var_1+=1

        var_2 = len(new_column) - var_1;

        #print(var_1, var_2)


        positive = [];
        negative = [];

        for x,y in list(zip(new_column, train.Loan_Status)):
            if y == 1:
                positive.append(x)
            if y == 0:
                negative.append(x)


        var_1_positive = 0;
        var_2_positive = 0;

        for x in positive:
            if x == 1:
                var_1_positive+=1
            if x == 0:
                var_2_positive+=1

        #print(var_1_positive, var_2_positive)

        var_1_negative = 0;
        var_2_negative = 0;

        for x in negative:
            if x == 1:
                var_1_negative+=1
            if x == 0:
                var_2_negative+=1

        #print(var_1_negative, var_2_negative)

        percent_positive = len(positive)/len(new_column) * 100
        percent_negative = len(negative)/len(new_column) * 100

        percent_var_2_positive = var_2_positive/var_2 * 100
        percent_var_1_positive = var_1_positive/var_1 * 100

        #print(percent_var_2_positive, percent_var_1_positive)

        percent_var_2_negative = var_2_negative/var_2 * 100
        percent_var_1_negative = var_1_negative/var_1 * 100

        #print(percent_var_2_negative,percent_var_1_negative)

        Info_array = np.array([percent_var_2_positive,percent_var_1_positive,percent_var_2_negative,percent_var_1_negative])

        return Info_array
    
    if len(new_column.unique()) == 3:
        var_1 = 0
        var_2 = 0
        var_3 = 0
        
        for x in new_column:
            if x == 0:
                var_1+=1
            if x == 1:
                var_2+=1
            if x == 2:
                var_3+=1
        
        positive = []
        negative = []
        
        for x,y in list(zip(new_column, train.Loan_Status)):
            if y == 1:
                positive.append(x)
            else :
                negative.append(x)
                
        
        var_1_positive = 0
        var_2_positive = 0
        var_3_positive = 0
        
        for x in positive:
            if x == 0:
                var_1_positive+=1
            if x == 1:
                var_2_positive+=1
            if x == 2:
                var_3_positive+=1
                
        
        var_1_negative = 0
        var_2_negative = 0
        var_3_negative = 0
        
        for x in negative:
            if x == 0:
                var_1_negative+=1
            if x == 1:
                var_2_negative+=1
            if x == 2:
                var_3_negative+=1
                
        percent_positive = len(positive)/len(new_column) * 100
        percent_negative = len(negative)/len(new_column) * 100

        percent_var_1_positive = var_1_positive/var_1 * 100
        percent_var_2_positive = var_2_positive/var_2 * 100
        percent_var_3_positive = var_3_positive/var_3 * 100
        
        percent_var_1_negative = var_1_negative/var_1 * 100
        percent_var_2_negative = var_2_negative/var_2 * 100
        percent_var_3_negative = var_3_negative/var_3 * 100
        
        Info_array = np.array([percent_var_1_positive, percent_var_2_positive, percent_var_3_positive, percent_var_1_negative, percent_var_2_negative, percent_var_3_negative])
             
        return Info_array
    
    if len(new_column.unique()) == 4:
        var_1 = 0
        var_2 = 0
        var_3 = 0
        var_4 = 0
        
        for x in new_column:
            if x == 0:
                var_1+=1
            if x == 1:
                var_2+=1
            if x == 2:
                var_3+=1
            if x == 3:
                var_4+=1
        
        
        positive = []
        negative = []
        
        for x,y in list(zip(new_column, train.Loan_Status)):
            if y == 1:
                positive.append(x)
            else :
                negative.append(x)
                
        
        var_1_positive = 0
        var_2_positive = 0
        var_3_positive = 0
        var_4_positive = 0
        
        for x in positive:
            if x == 0:
                var_1_positive+=1
            if x == 1:
                var_2_positive+=1
            if x == 2:
                var_3_positive+=1
            if x == 3:
                var_4_positive+=1
                
        
        var_1_negative = 0
        var_2_negative = 0
        var_3_negative = 0
        var_4_negative = 0
        
        for x in negative:
            if x == 0:
                var_1_negative+=1
            if x == 1:
                var_2_negative+=1
            if x == 2:
                var_3_negative+=1
            if x == 3:
                var_4_negative+=1
                
        percent_positive = len(positive)/len(new_column) * 100
        percent_negative = len(negative)/len(new_column) * 100

        percent_var_1_positive = var_1_positive/var_1 * 100
        percent_var_2_positive = var_2_positive/var_2 * 100
        percent_var_3_positive = var_3_positive/var_3 * 100
        percent_var_4_positive = var_4_positive/var_4 * 100
        
        percent_var_1_negative = var_1_negative/var_1 * 100
        percent_var_2_negative = var_2_negative/var_2 * 100
        percent_var_3_negative = var_3_negative/var_3 * 100
        percent_var_4_negative = var_4_negative/var_4 * 100
        
        Info_array = np.array([percent_var_1_positive, percent_var_2_positive, percent_var_3_positive,percent_var_4_positive, percent_var_1_negative, percent_var_2_negative, percent_var_3_negative, percent_var_4_negative])
             
        return Info_array
```

```python
# Bar Graph to show the relation between different Genders and Loan approval 

def graphing(label, title, xlabel, Info_array):

    ''' index = np.arange(len(Info_array))
    fig_size =(12,12) 
    bar = plt.bar(index, Info_array)
    bar[0].set_color('g')
    bar[1].set_color('g')
    plt.ylabel('Percentage')
    plt.xlabel(xlabel)
    plt.xticks(index,label)
    plt.title(title)
    plt.rcParams["figure.figsize"] = fig_size
    plt.show() '''
    
    positive = Info_array[0:len(label)]
    negative = Info_array[len(label):]
    index = np.arange(len(positive))
    width = 0.3
    
    fig, ax = plt.subplots()
    bar_1 = ax.bar(index, positive, width, color= 'g')
    bar_2 = ax.bar(index + width, negative, width, color= 'r')
    
    fig_size = plt.rcParams["figure.figsize"]
    fig_size[0] = 12
    fig_size[1] = 12
    plt.rcParams["figure.figsize"] = fig_size
    
    ax.set_ylabel('Percentage')
    ax.set_xlabel(xlabel)
    ax.set_xticks(index + width/2)
    ax.set_xticklabels((label))
    ax.set_title(title)
    ax.legend((bar_1[0], bar_2[0]),('Approved', 'Disapproved'))
    
    
    plt.show()
    
```

```python
def calculator_cont(new_column):
    mean = sat.mean(new_column)
    median = sat.median(new_column)
    mode = sat.mode(new_column)
    std = sat.stdev(new_column)
    
    plt.scatter(new_column, train.Loan_Status)

    
    positive = []
    negative = []
    
    for x,y in list(zip(new_column, train.Loan_Status)):
        if y == 1:
            positive.append(x)
        else:
            negative.append(x)
            
    mean_positive = sat.mean(positive)
    mean_negative = sat.mean(negative)
    
    median_positive = sat.median(positive)
    median_negative = sat.median(negative)
    
    mode_positive = sat.mode(positive)
    mode_negative = sat.mode(negative)
    
    std_positive = sat.stdev(positive)
    std_negative = sat.stdev(negative)
    
    Info_array = np.array([mean_positive, mean_negative, median_positive, median_negative, mode_positive, mode_negative, std_positive, std_negative])
    plt.show()
    return Info_array
```

```python
def graph_cont(label, title, xlabel, Info):
    mean = Info[0:2]
    median = Info[2:4]
    mode = Info[4:6]
    std = Info[6:]
    
    index = np.arange(2)
    width = 0.2
    
    fig, ax = plt.subplots()
    bar_1 = ax.bar(index, mean, width, color= 'g')
    bar_2 = ax.bar(index + width, median, width, color= 'y')
    bar_3 = ax.bar(index + 2*width, mode, width, color= 'b')
    bar_4 = ax.bar(index + 3*width, std, width, color= 'r')
    
    fig_size = plt.rcParams["figure.figsize"]
    fig_size[0] = 14
    fig_size[1] = 14
    plt.rcParams["figure.figsize"] = fig_size
    
    ax.set_ylabel('value')
    ax.set_xlabel(xlabel)
    ax.set_xticks(index + width/2)
    ax.set_xticklabels(label)
    ax.set_title(title)
    ax.legend((bar_1[0], bar_2[0], bar_3[0], bar_4[0]),('Mean', 'Median', 'Mode', 'Std'))
        
    plt.show()
```

```python
Info = calculator_cont(train.ApplicantIncome)
label = ['Approved', 'Not Approved']
xlabel = 'Loan Status'
title = 'Relation betwen Loan Status and Applicant\'s Income'

graph_cont(label, title, xlabel, Info)
```

```python
Info = calculator_cont(train.CoapplicantIncome)
label = ['Approved', 'Not Approved']
xlabel = 'Loan Status'
title = 'Relation betwen Loan Status and Applicant\'s Income'

graph_cont(label, title, xlabel, Info)
```
